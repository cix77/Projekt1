```python
import pandas as pd
import requests

base_url = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv"

params = {
    "$where": "complaint_type='Noise - Vehicle'",
    "$limit": 1000
}

response = requests.get(base_url, params=params)

# 2. Prüfen, ob der Download geklappt hat
if response.status_code == 200:
    # Über ein Text-Stream-Objekt direkt in Pandas einlesen
    from io import StringIO
    df = pd.read_csv(StringIO(response.text))
    print(f"Erfolgreich geladen! Datensatz-Form: {df.shape}")
    print(df.head())
else:
    print(f"Fehler beim Download: Status Code {response.status_code}")
```
