```python
import pandas as pd

url = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv?$where=complaint_type='Noise - Vehicle'&$limit=1000"
df = pd.read_csv(url)
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from io import StringIO

base_url = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv"
params = {
    "$where": "complaint_type='Noise - Vehicle'",
    "$limit": 1000
}

response = requests.get(base_url, params=params)

if response.status_code == 200:
    df = pd.read_csv(StringIO(response.text))
    print("Successfully loaded! Dataset shape:", df.shape)
    print(df[['latitude', 'longitude']].head())
else:
    print("Error loading data. Status Code:", response.status_code)
```

```python
## Map 1: Simple Scatter Plot of NYC Complaints
Now we plot the longitude and latitude of all complaints. 
This will draw a rough map of New York City based only on the locations of the noise complaints.
We save this map as an SVG file for our website.
```

```python
# Clean the data: remove rows that don't have coordinates
df_clean = df.dropna(subset=['latitude', 'longitude'])

# Create a figure
plt.figure(figsize=(10, 12))

# Plot the coordinates
# alpha=0.3 makes the dots slightly transparent so we can see density
# s=5 makes the dots small
plt.scatter(df_clean['longitude'], df_clean['latitude'], alpha=0.3, s=5, color='#2c3e6b')

# Remove the axis ticks to make it look like a real map
plt.axis('off')
plt.title('311 Noise Complaints Map - New York City')

# Save as SVG in your project folder
plt.savefig("noisemap.svg", format="svg", bbox_inches='tight')
plt.show()
```

```python
## Map 2: Density Heatmap of Manhattan Hotspots
To find the absolute loudest hotspots, we filter our data specifically for Manhattan coordinates.
Instead of Seaborn, we use matplotlib's built-in "hexbin" function. 
This groups the dots into small hexagons and colors them by frequency. Red areas show where the most complaints are.
We save this map as a PNG.
```

```python
# Filter coordinates for Manhattan (zooming in)
manhattan_df = df_clean[
    (df_clean['latitude'] > 40.70) & (df_clean['latitude'] < 40.88) & 
    (df_clean['longitude'] > -74.02) & (df_clean['longitude'] < -73.90)
]

# Create the figure
plt.figure(figsize=(8, 12))

# Draw the hexbin map
# gridsize=50 defines how detailed the hexagons are
# cmap='Reds' colors the dense areas dark red
# mincnt=1 ensures we don't draw hexagons where there are 0 complaints
plt.hexbin(
    manhattan_df['longitude'], 
    manhattan_df['latitude'], 
    gridsize=50, 
    cmap='Reds', 
    mincnt=1
)

# Remove axis and add title
plt.axis('off')
plt.title('Manhattan Noise Hotspots (Washington Heights Area)')

# Save as PNG in your project folder
plt.savefig("heatmap33.png", format="png", dpi=150, bbox_inches='tight')
plt.show()
```
